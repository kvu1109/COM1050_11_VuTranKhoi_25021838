```python
#W4A2
def check_prime(n):
    if n <= 1:
        return False
    else:
        for i in range(2, int(n**0.5) + 1):
            if n % i == 0:
                return False
        return True
while True:
    try:
        n = int(input())
        if n <= 0:
            continue
        else:
            print(check_prime(n))
            break
    except ValueError:
        print("Hãy nhập số")
```

```python
#W4A3
def factorial(n):
    k = 1
    f = 1
    while k <= n:
        f *= k
        k += 1
    return f
n = int(input())
print(factorial(n))
```

```python
#W4A4
def count_digit(n):
    count = 0
    if n < 0:
        n = -n
    while n >= 1:
        count += 1
        n /= 10
    return count
n = int(input())
print(count_digit(n))
```

```python
#W4A6
def prime(n):
    if n <= 1:
        return None
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return None
    return n
    
a, b = map(int, input().split())
prime_sum = 0
for i in range(a, b+1):
    if i == prime(i):
        prime_sum += i
print(prime_sum)
```

```python
#W4A9
def distinct_digits(n):
    s = str(n)
    set = []
    for i in s: 
        if i in set:
            return None
        set += i
    else: 
        return n

def squares(n):
    if n**0.5 == int(n**0.5):
        return n
    else:
        return None
n = int(input())
for i in range(1, n+1):
    if distinct_digits(squares(i)) != None:
        print(distinct_digits(squares(i)), end=" ")    
```

```python
#W4A10
def collatz(n):
    if n == 1:
        return 0
    else:
        step_count = 1
        while n != 1:
            step_count += 1
            if n % 2 == 0:
                n /= 2
            else:
                n = n*3 + 1
    return step_count
n = int(input())
min_num = n
max_step = 0
set = []
for i in range(n+1, 1, -1):
    if collatz(i) >= max_step and i <= min_num:
        max_step = collatz(i)
        min_num = i
print(min_num, max_step)
```

```python
#W4A26
def fibonacci(n):
    if n <= 0:
        return None
    elif n <= 2:
        return 1
    else:
        return fibonacci(n-2) + fibonacci(n-1)
n = int(input())
i = 1
while fibonacci(i) <= n:
    i += 1
print(fibonacci(i-1))
```
