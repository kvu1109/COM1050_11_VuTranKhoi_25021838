```python
#W4A1 [SumOfNumBers]
def sum_n(n):
  sum = 0
  for i in range(1,n+1):
    sum += i
  return sum

n = int(input())
print(sum_n(n))
```

```python
#W4A2 [PrimeNumber]
def check_prime(n):
  if n < 2:
    return False
  else:
    for i in range(2, int(n**0.5) + 1):
      if n % i == 0:
        return False
    return True

prime = False
while not prime:
  try:
      n = int(input())
      if check_prime(n):
        print("True")
        prime = True
      else:
        print("False")
  except ValueError:
    continue
```

```python
#W4A3 [Factorial]
def factorial(n):
  f = 1
  for i in range(1, n+1):
    f *= i
  return f

while True:
  try:
    n = int(input())
    if n >= 100 or n <=0:
      continue
    else:
      print(factorial(n))
      break
  except ValueError:
    continue
```

```python
#W4A4 [CountingDigits]
def count_digits(n):
  count = 0
  if n == 0:
    return 1
  if n < 0:
    n = n*-1
    while n > 0:
      n = n // 10
      count += 1
    return count
  else:
    while n > 0:
      n = n // 10
      count += 1
    return count

n = int(input())
print(count_digits(n))
```

```python
#W4A6 [SumPrimesInRange]
def check_prime(n):
  if n < 2:
    return 0
  else:
    for i in range(2, int(n**0.5) + 1):
      if n % i == 0:
        return 0
    return n

a, b = map(int, input().split())
sum = 0
for j in range(a, b+1):
    if j == check_prime(j):
        sum += j
print(sum)
```
