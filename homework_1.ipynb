```python
#Bài 2
while True:
    try:
        a, b = map(float, input().split())
        if a > b or a < 0 or b<0:
            print("Nhập lại dữ liệu dương và nhập chiều rộng trước")
            continue
        if a < b and a > 0 and b > 0:
            S_dat = a*b
            S_kvc = ((a/2)**2)*3.14
            S_cay = round(S_dat - S_kvc, 2)
            print(S_cay)
            break
    except ValueError:
        print("Hãy nhập lại số hợp lệ")
```

```python
#Bài 3
while True:
    x = str (input())
    if len(x) != 1:
        print("Hãy nhập 1 kí tự chữ cái")
        continue
    try:
        n = ord(x)
        if n >= 65 and n <= 90:
            print(chr(n+32))
            break
        elif n >=97 and n <= 125:
            print(chr(n-32))
            break
        else:
            print("Hãy nhập 1 kí tự chữ cái")
            continue
    except ValueError:
        print("Hãy nhập dữ liệu dạng xâu")
```

```python
#Bài 4
while True:
    x = str(input())
    if len(x) != 1:
        print(f"{x} không phải kí tự alphabet")
        continue
    try:
        n = ord(x)
        if 65 <= n <= 90 or 97 <= n <= 125:
            print(f"{x} là kí tự alphabet")
            break
        else:
            print(f"{x} không phải kí tự alphabet")
            break
    except ValueError:
        print("Hãy nhập một kí tự")
```

```python
#Bài 5
while True:
    x = str(input())
    if len(x) != 1:
        print("Hãy nhập 1 kí tự in hoa")
        continue
    try:
        n = ord(x)
        if 65 < n <= 90:
            print(chr(n +31))
            break
        elif n == 65:
            print(f"{x} là kí tự đặc biệt không có chữ in thường liền trước")
            break
        else:
            print("Hãy điền 1 kí tự in hoa")
            continue
    except ValueError:
        print("Hãy nhập 1 kí tự in hoa hợp lệ")
```

```python
#Bài 6
import math

while True:
    try:
        values = input().split()
        if len(values) != 3:
            print("Hãy nhập đúng 3 số")
            continue
        a, b, c = map(float, values)
        if a + b <= c or b + c <= a or c + a <= b or a <= 0 or b <= 0 or c <= 0:
            print(f"{a} {b} {c} không phải độ dài 3 cạnh 1 tam giác. Hãy nhập lại")
            continue
        else:
            p = (a+b+c)/2
            s = p*(p-a)*(p-b)*(p-c)
            S = math.sqrt(s)
            final = round(S, 1)
            print(final)
            break
    except ValueError:
        print("Hãy nhập 3 số hợp lệ")
            
```

```python
#Bài 8
while True:
    try:
        tch = str(input("Tên chủ hộ: "))
        cstt = int(input("Chỉ số tháng trước: "))
        cstn = int(input("Chỉ số tháng này: "))
        if cstt > cstn or cstt < 0 or cstn < 0:
            print("Hãy nhập lại giá trị hợp lệ")
            continue
        kwh = cstn - cstt
        tiers = [
            (50, 1984),   # 0-50 kWh
            (100, 2050),  # 51-100 kWh
            (200, 2380),  # 101-200 kWh
            (300, 2998),  # 201-300 kWh
            (400, 3350),  # 301-400 kWh
            (float('inf'), 3460)  # 401+ kWh
        ]

        total_cost = 0
        remaining_kwh = kwh
        prev_limit = 0

        for limit, price in tiers:
            if remaining_kwh <= 0:
                break
            # Số kWh tiêu thụ trong bậc hiện tại
            kwh_in_tier = min(remaining_kwh, limit - prev_limit)
            total_cost += kwh_in_tier * price
            remaining_kwh -= kwh_in_tier
            prev_limit = limit

        total_cost_final = round(total_cost*108/100)
        print(f"Họ và tên: {tch}")
        print(f"Tiền phải trả là: {total_cost_final}")
        break
    except ValueError:
        print("Hãy nhập chỉ số hợp lệ")
```
