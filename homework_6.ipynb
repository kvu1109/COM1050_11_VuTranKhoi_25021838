```python
#W6A1
def list_rut_gon(l):
    result = []
    seen = set()
    for i in l:
        if i not in seen:
            result.append(i)
            seen.add(i)
    return result
l = list(map(int, input().split()))
print(list_rut_gon(l))
```

```python
#W6A2
def tong_tich_luy(l):
    for i in range(len(l)):
        if i == 0:
            l[i] = l[i]
        else:
            l[i] += l[i-1]
    return l
l = list(map(int, input().split()))
print(tong_tich_luy(l))
```

```python
#W6A3
def xoay(l, k):
    a = 0
    for i in range(k):
        a = l[0]
        for j in range(len(l)):
            if j == len(l) - 1:
                l[j] = a
            else:
                l[j] = l[j+1]
    return tuple(l)
l = list(map(int, input().split()))
k = int(input())
print(xoay(l, k))
```

```python
#W6A5
def dem_so(l):
    result = {}
    po = 0
    ne = 0
    zeros = 0
    for i in range(len(l)):
        if l[i] > 0:
            po += 1
        elif l[i] < 0:
            ne += 1
        else:
            zeros += 1
    result['positives'] = po
    result['negatives'] = ne
    result['zeros'] = zeros
    return result
l = list(map(int, input().split()))
print(dem_so(l))
```

```python
#W6A8
def dem_phan_tu(l):
    result = {}
    for i in range(len(l)):
        a = l[i]
        if a in result:
            result[a] += 1
        else:
            result[a] = 1
    return result
l = list(map(str, input().split()))
print(dem_phan_tu(l))
```

```python
#W6A9
def gop_dict(l1, l2):
    result = {}
    for key in l1:
        if key in l2:
            result[key] = l1[key] + l2[key] 
        else:
            result[key] = l1[key]           
    for key in l2:
        if key not in result:
            result[key] = l2[key]
    return {key: result[key] for key in sorted(result)}
l1 = eval(input())
l2 = eval(input())
print(gop_dict(l1, l2))
```

```python
#W6A10:
def tong(l, k):
    result = set()
    for i in range(len(l)):
        for j in range(i + 1, len(l)):
            if l[i] + l[j] == k:
                result.add((l[i], l[j]))
    return sorted(list(result))
l = list(map(int, input().split()))
k = int(input())
print(tong(l, k))
```

```python
#W6A12
def dem(l):
    dic = {}
    for i in l:
        if i not in dic:
            dic[i] = 1
        else:
            dic[i] += 1
    return dic
def phan_tu_xuat_hien_nhieu_nhat(dic):
    value = 0
    for i in dic:
        if dic[i] >= value:
            value = dic[i]
    result_list = []
    for i in dic:
        if dic[i] == value:
            result_list.append(i)
    return min(result_list)
d = list(map(int, input().split()))
print(phan_tu_xuat_hien_nhieu_nhat(dem(d)))
```

```python
#W6A13
def reverse_dict(dic):
    r_dict = {}
    for i in range(0, len(dic), 2):
        r_dict[dic[i+1]] = dic[i]
    return r_dict
d = input().split()
print(reverse_dict(d))
```

```python
#W6A14
def giao_list(l1, l2):
    giao = set()
    for i in l1:
        if i in l2:
            giao.add(i)
    result = list(giao)
    result.sort()
    return result
l1 = list(map(int, input().split()))
l2 = list(map(int, input().split()))
print(giao_list(l1, l2))
```

```python
#W6A15
def bigger_than_k(dic, k):
    result = {}
    dictionary = {}
    for i in range(0, len(dic), 2):
        dictionary[dic[i]] = int(dic[i+1])
    for i in dictionary:
        if dictionary[i] > k:
            result[i] = dictionary[i]
    return result
d = list(map(str, input().split()))
k = int(input())
print(bigger_than_k(d, k))
```

```python
#W6A17
n = int(input("Nhập cỡ của ma trận"))
matrix = []
for i in range(n):
    while True:
        row = list(map(int, input(f"Nhập hàng {i+1}: ").split()))
        if len(row) != n:
            print(f"Hãy nhập đúng {n} phần tử")
            continue
        matrix.append(row)
        break
index_chinh = 0
index_phu = n - 1
cheo_chinh = []
cheo_phu = []
for row in matrix:
    cheo_chinh.append(row[index_chinh])
    cheo_phu.append(row[index_phu])
    index_chinh += 1
    index_phu -= 1
print(f"Đường chéo chính là {cheo_chinh}")
print(f"Đường chéo phụ là {cheo_phu}")
```

```python
#W6A18:
n, m, k = map(int, input("Nhập số hàng, số cột, và cột/hàng cần tính tổng: ").split())
matrix = []
for i in range(n):
    while True: 
        row = list(map(int, input(f"Nhập hàng {i+1}: ").split()))
        if len(row) != m:
            print(f"Hãy nhập đúng {m} phần tử")
            continue  
        matrix.append(row)
        break 
sum_cot = 0
sum_hang = 0
#Tính sum_cot:
for i in range(len(matrix)):
    for j in range(len(matrix[i])):
        if j+1 == k:
            sum_cot += matrix[i][j]
#Tính sum_hang:
for i in range(len(matrix)):
    if i + 1 == k:
        sum_hang = sum(matrix[i])
if k > n and k > m:
    print("Không tính được")
elif k > n and k <= m:
    print(f"Cột {k} là: {[matrix[i][k-1] for i in range(n)]}, có tổng là: {sum_cot}")
elif k <= n and k > m:
    print(f"Hàng {k} là: {matrix[k-1]}, có tổng là: {sum_hang}")
else:
    print(f"Hàng {k} là: {matrix[k-1]}, có tổng là: {sum_hang}")
    print(f"Cột {k} là: {[matrix[i][k-1] for i in range(n)]}, có tổng là: {sum_cot}")
```

```python
#W6A19
def position_max_of_list(l):
    max_result = l[0]
    index_max_result = 0
    for i in range(len(l)):
        if l[i] >= max_result:
            max_result = l[i]
            index_max_result = i + 1
    return index_max_result
l = list(map(int, input().split()))
print(position_max_of_list(l))
```

```python
#W6A20
def find_value(l, k):
    result = -1
    for i in range(len(l) - 1, -1, -1):
        if l[i] == k:
            result = i + 1
    return result
l = list(map(int, input().split()))
k = int(input())
print(find_value(l, k))
```

```python
#W6A21
def anh_xa(l):
    dic = {}
    for i in range(len(l)):
        if l[i] not in dic:
            dic[l[i]] = [i]
        else:
            dic[l[i]].append(i)
    for i in dic:
        dic[i] = tuple(dic[i])
    return dic
l = list(map(str, input().split()))
print(anh_xa(l))
```
