```python
#W8A1
class Cylinder:
    def __init__(self, radius, height):
        self.radius = radius
        self.height = height

    def calculate_cylinder(self):
        r = self.radius
        h = self.height
        pi = 3.14
        
        a = 2*pi*r*h + 2*pi*(r**2)
        v = pi*(r**2)*h
        print(round(a, 2), round(v, 2))
r, h = map(float, input().split())
c = Cylinder(r, h)
c.calculate_cylinder()
```

```python
#W8A2
class Date:
    def __init__(self, day, month, year):
        self.day = day
        self.month = month
        self.year = year
        self.days_in_month = [0, 31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]

    def isleapyear(self):
        if self.year % 400 == 0 or self.year % 4 == 0 and self.year % 100 != 0:
            return True
        else:
            return False
    
    def valid_date(self):
        if self.isleapyear():
            self.days_in_month[2] = 29
        if self.day <= 0 or self.day > self.days_in_month[self.month] or self.month <= 0 or self.month > 12 or self.year <= 0:
            return False
        else:
            return True

    def next_date(self):
        if self.valid_date():
            self.day += 1
            if self.day > self.days_in_month[self.month]:
                self.day = 1
                self.month += 1
                if self.month > 12:
                    self.month = 1
                    self.year += 1
            return f"{self.day:02}/{self.month:02}/{self.year:02}"
        else:
            return 'INVALID'
d, m, y = map(int, input().split("/"))
date = Date(d, m, y)
print(date.next_date())
```

```python
#W8A3
class Point:
    def __init__(self, x_val, y_val):
        self.x = x_val
        self.y = y_val

    def goc_toa_do(self):
        if self.x == 0 and self.y == 0:
            print(f"Điểm ({self.x}, {self.y}) là gốc tọa độ")
        else:
            print(f"Điểm ({self.x}, {self.y}) không là gốc tọa độ")

    def truc_tung(self):
        if self.x == 0:
            print(f"Điểm ({self.x}, {self.y}) nằm trên trục tung")
        else:
            print(f"Điểm ({self.x}, {self.y}) không nằm trên trục tung")

    def truc_hoanh(self):
        if self.y == 0:
            print(f"Điểm ({self.x}, {self.y}) nằm trên trục hoành")
        else:
            print(f"Điểm ({self.x}, {self.y}) không nằm trên trục hoành")

    def khoang_cach(self):
        d = ((self.x)**2 + (self.y)**2)**0.5
        print(f"Khoảng cách của điểm ({self.x}, {self.y}) tới gốc tọa độ là {round(d, 2)}")
x, y = map(int, input().split())
point = Point(x, y)
point.goc_toa_do()
point.truc_tung()
point.truc_hoanh()
point.khoang_cach()
```

```python
#W8A5
class ShoppingCart:
    def __init__(self):
        self.items = []

    def add_item(self, name, price):
        product = {"name": name, "price": price}
        self.items.append(product)
        print(f"Đã thêm '{name}' vào giỏ hàng.")

    def remove_item(self, name):
        for item in self.items:
            if item['name'].lower() == name.lower():
                self.items.remove(item)
                print(f"Đã xóa '{name}' khỏi giỏ hàng.")
                return
        print(f"Không tìm thấy sản phẩm '{name}' để xóa.")

    def is_empty(self):
        return len(self.items) == 0

    def total_price(self):
        return sum(item['price'] for item in self.items)

    def display_items(self):
        if self.is_empty():
            print("Giỏ hàng hiện đang rỗng.")
        else:
            print("--- CÁC SẢN PHẨM TRONG GIỎ ---")
            for i, item in enumerate(self.items, 1):
                print(f"{i}. {item['name']} - {item['price']:,} VNĐ")
            print(f"Tổng cộng: {self.total_price():,} VNĐ")

    def clear_cart(self):
        self.items = []
        print("Đã làm trống giỏ hàng.")

# --- CHƯƠNG TRÌNH CHÍNH (MENU) ---
cart = ShoppingCart()

while True:
    print("\n========= MENU GIỎ HÀNG =========")
    print("2. Thêm sản phẩm")
    print("3. Xóa sản phẩm theo tên")
    print("4. Kiểm tra giỏ hàng rỗng")
    print("5. Tính tổng tiền")
    print("6. Hiển thị giỏ hàng")
    print("7. Xóa toàn bộ giỏ hàng")
    
    choice = input("Nhập lựa chọn của bạn (2-7): ")

    if choice == '2':
        name = input("Nhập tên sản phẩm: ")
        price = float(input("Nhập giá sản phẩm: "))
        cart.add_item(name, price)
    elif choice == '3':
        name = input("Nhập tên sản phẩm cần xóa: ")
        cart.remove_item(name)
    elif choice == '4':
        if cart.is_empty():
            print("Giỏ hàng đang rỗng.")
        else:
            print(f"Giỏ hàng đang có {len(cart.items)} sản phẩm.")
    elif choice == '5':
        print(f"Tổng số tiền: {cart.total_price():,} VNĐ")
    elif choice == '6':
        cart.display_items()
    elif choice == '7':
        cart.clear_cart()
    else:
        print("Lựa chọn không hợp lệ!")

    # Hỏi thoát chương trình
    exit_confirm = input("\nCó muốn thoát khỏi chương trình không? (yes/no): ").lower()
    if exit_confirm == "yes":
        print("Cảm ơn bạn đã sử dụng dịch vụ!")
        break
```

```python
#W8A6
class Rectangle:
    def __init__(self, w, h):
        self.w = w
        self.h = h

    def area(self):
        print(round((self.w * self.h), 2))

    def perimeter(self):
        print(round((2*(self.w + self.h)), 2))

    def scale(self, k):
        self.w *= k
        self.h *= k
        self.area()
        self.perimeter()

w, h, k = map(float, input().split())
rec = Rectangle(w, h)
rec.scale(k)
```

```python
#W8A7
def gcd(a, b):
    if b == 0:
        return a
    else:
        return gcd(b, a%b)
        
class Fraction:
    def __init__(self, num, den):
        if den == 0:
            raise ValueError("Mẫu số không được bằng 0")
        elif den < 0:
            num = -num
            den = -den
        gcd_val = gcd(num, den)
        self.num = num // gcd_val
        self.den = den // gcd_val
        
    def __add__(self, other):
        new_num = self.num * other.den + other.num * self.den
        new_den = self.den * other.den
        return Fraction(new_num, new_den)
        
    def __sub__(self, other):
        new_num = self.num * other.den - other.num * self.den
        new_den = self.den * other.den
        return Fraction(new_num, new_den)
        
    def __mul__(self, other):
        new_num = self.num * other.num
        new_den = self.den * other.den
        return Fraction(new_num, new_den)
        
    def __truediv__(self, other):
        if other.num == 0:
            raise ValueError("Chia cho phân số 0")
        new_num = self.num * other.den
        new_den = self.den * other.num
        return Fraction(new_num, new_den)

    def __str__(self):
        return f"{self.num}/{self.den}"

data = input().split()
a = int(data[0])
b = int(data[1])
op = data[2]
c = int(data[3])
d = int(data[4])

f1 = Fraction(a, b)
f2 = Fraction(c, d)

if op == "+":
    result = f1 + f2
elif op == "-":
    result = f1 - f2
elif op == "*":
    result = f1 * f2
elif op == "/":
    result = f1 / f2
else:
    print("Toán tử không hợp lệ")
    exit()

print(result)
        
```

```python
#W8A8
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = float(balance)

    def deposit(self, x):
        self.balance += float(x)

    def withdraw(self, x):
        if float(x) > self.balance:
            pass
        else:
            self.balance -= float(x)
    
    def __str__(self):
        return f"{self.owner}: {self.balance:.2f}"

#main
data = input().split()
o = data[0]
b = data[1]

account = BankAccount(o, b)

n = int(input())

while n > 0:
    trade = input().split()
    if trade[0] == "DEPOSIT":
        account.deposit(trade[1])
    else:
        account.withdraw(trade[1])
    n -= 1
print(account)
```

```python
#W8A9
class Student:
    def __init__(self, name):
        self.name = name
        self.all_scores = []
        
    def add_score(self, subject, score):
        self.all_scores.append(float(score))
        
    def avg(self):
        avg = sum(self.all_scores)/len(self.all_scores)
        return avg
        
    def rank(self):
        if self.avg() >= 8:
            return "Excellent"
        elif self.avg() >= 6.5:
            return "Good"
        elif self.avg() >= 5:
            return "Average"
        else:
            return "Poor"
    def __str__(self):
        self.avg()
        self.rank()
        return f"{self.name} {self.avg():.2f} {self.rank()}"

name = Student(input())
n = int(input())
while n > 0:
    data = input().split()
    name.add_score(data[0], data[1])
    n -= 1
print(name)
```

```python
#W8A10
class Book:
    def __init__(self, title, author, year):
        self.title = title
        self.author = author
        self.year = int(year)

class Library():
    def __init__(self):
        self.book_list = []

    def add(self, book):
        self.book_list.append(book)

    def count_by_author(self, name):
        count = 0
        for i in self.book_list:
            if i.author == name:
                count += 1
        return count

    def find_by_year(self, y):
        count = 0
        for i in self.book_list:
            if i.year == int(y):
                count += 1
        return count

n = int(input())
big_data = Library()
while n > 0:
    x = input().split()
    if x[0] == "ADD":
        parts = x[1].split(";")
        data = Book(parts[0], parts[1], parts[2])
        big_data.add(data)
    elif x[0] == "COUNT":
        print(big_data.count_by_author(x[1]))
    else:
        print(big_data.find_by_year(x[1]))
    n -= 1
```

```python
#W8A11
import math

class Vector:
    def __init__(self, x, y, z):
        self.x, self.y, self.z = x, y, z

    def __sub__(self, other):
        return Vector(self.x - other.x, self.y - other.y, self.z - other.z)

    def dot(self, other):
        return self.x * other.x + self.y * other.y + self.z * other.z

    def cross(self, other):
        return Vector(self.y * other.z - self.z * other.y,
                      self.z * other.x - self.x * other.z,
                      self.x * other.y - self.y * other.x)

    def length(self):
        return math.sqrt(self.x**2 + self.y**2 + self.z**2)

# Nhập dữ liệu cho 4 điểm A, B, C, D
points = []
for point in range(4):
    points.append(Vector(*map(float, input().split())))

A, B, C, D = points

# 1. Tính các vector tạo thành mặt phẳng
AB = B - A
BC = C - B
CD = D - C

# 2. Tính vector pháp tuyến của hai mặt phẳng (ABC) và (BCD)
# Vector pháp tuyến = Tích có hướng của 2 vector nằm trong mặt phẳng
n1 = AB.cross(BC)
n2 = BC.cross(CD)

# 3. Tính góc giữa 2 vector pháp tuyến
# cos(phi) = (n1 . n2) / (|n1| * |n2|)
cos_phi = n1.dot(n2) / (n1.length() * n2.length())

# Đảm bảo giá trị nằm trong [-1, 1] để tránh lỗi tính toán dấu phẩy động
cos_phi = max(min(cos_phi, 1.0), -1.0)

# Tính độ
phi_rad = math.acos(cos_phi)
phi_deg = math.degrees(phi_rad)

# In kết quả
print(f"{phi_deg:.2f}")
```

```python
#W8A12
class ComplexNumber:
    def __init__(self, initRealPart, initImaginaryPart):
        self.realPart = float(initRealPart)
        self.imaginaryPart = float(initImaginaryPart)

    # Hàm nhân số phức hiện tại với một số phức khác
    def multiply(self, other):
        # (a+bi)(c+di) = (ac-bd) + (ad+bc)i
        new_real = (self.realPart * other.realPart) - (self.imaginaryPart * other.imaginaryPart)
        new_imag = (self.realPart * other.imaginaryPart) + (self.imaginaryPart * other.realPart)
        return ComplexNumber(new_real, new_imag)

    def __str__(self):
        return f"{self.realPart:.2f} {self.imaginaryPart:.2f}"

# Nhập số phức thứ nhất
r1, i1 = map(float, input().split())
c1 = ComplexNumber(r1, i1)

# Nhập số phức thứ hai
r2, i2 = map(float, input().split())
c2 = ComplexNumber(r2, i2)

# Tính tích
result = c1.multiply(c2)

print(f"{result.realPart} {result.imaginaryPart}")
```

```python
#W8A13
import math

class Triangle:
    def __init__(self, initEdge1, initEdge2, initEdge3):
        self.edge1 = initEdge1
        self.edge2 = initEdge2
        self.edge3 = initEdge3

    def getArea(self):
        a, b, c = self.edge1, self.edge2, self.edge3
        
        if (a + b > c) and (a + c > b) and (b + c > a) and (a > 0 and b > 0 and c > 0):
            p = (a + b + c) / 2
            area = math.sqrt(p * (p - a) * (p - b) * (p - c))
            return area
        else:
            return -1

try:
    data = list(map(int, input().split()))
    if len(data) == 3:
        a, b, c = data
        tri = Triangle(a, b, c)
        result = tri.getArea()
        
        if result == -1:
            print("invalid")
        else:
            print(f"{result:.2f}")
    else:
        print("invalid")
except ValueError:
    print("invalid")
```
