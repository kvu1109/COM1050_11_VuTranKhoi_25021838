```python
#W7A1
def binary_search(arr, value):
    left, right = 0, len(arr) - 1
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == value:
            return mid
        elif arr[mid] < value:
            left = mid + 1
        else:
            right = mid - 1
    return -1
l = list(map(int, input().split()))
t = int(input())
print(binary_search(l, t))
```

```python
#W7A2
def count_occurences(arr, x):
    count = 0
    for i in arr:
        if i == x:
            count += 1
    return count
l = list(map(int, input().split()))
t = int(input())
print(count_occurences(l, t))
```

```python
#W7A3
def bubble_sort(arr):
    for i in range(len(arr)):
        swap = False
        for j in range(len(arr) - i - 1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                swap = True
        if not swap:
            break
    return arr

def selection_sort(arr):
    for i in range(len(arr)):
        min_idx = i
        for j in range(i+1, len(arr)):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr

l = list(map(int, input().split()))
print(bubble_sort(l))
print(selection_sort(l))       
```

```python
#W7A4
def most_appear(arr):
    most_appear_count = 0
    most_appear_value = 0
    for i in range(len(arr) - 1, -1, -1):
        count = 1
        for j in range(i - 1, -1, -1):
            if arr[i] == arr[j]:
                count += 1
                if count >= most_appear_count:
                    most_appear_count = count
                    most_appear_value = arr[i]
    return f"{most_appear_value} xuat hien nhieu nhat, som nhat, {most_appear_count} lan"
l = list(map(int, input().split()))
print(most_appear(l))
```

```python
#W7A5
def dem_cap_bang_tong(arr, x):
    count = 0
    for i in range(len(arr)):
        for j in range(i+1, len(arr)):
            if x - arr[i] == arr[j]:
                count += 1
    return count
l = list(map(int, input().split()))
x = int(input())
print(dem_cap_bang_tong(l, x))
```
