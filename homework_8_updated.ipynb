```python
#W8A1
class Cylinder:
    def __init__(self, radius, height):
        self.radius = radius
        self.height = height

    def calculate_cylinder(self):
        r = self.radius
        h = self.height
        pi = 3.14
        
        a = 2*pi*r*h + 2*pi*(r**2)
        v = pi*(r**2)*h
        print(round(a, 2), round(v, 2))
r, h = map(float, input().split())
c = Cylinder(r, h)
c.calculate_cylinder()
```

```python
#W8A3
class Point:
    def __init__(self, x_val, y_val):
        self.x = x_val
        self.y = y_val

    def goc_toa_do(self):
        if self.x == 0 and self.y == 0:
            print(f"Điểm ({self.x}, {self.y}) là gốc tọa độ")
        else:
            print(f"Điểm ({self.x}, {self.y}) không là gốc tọa độ")

    def truc_tung(self):
        if self.x == 0:
            print(f"Điểm ({self.x}, {self.y}) nằm trên trục tung")
        else:
            print(f"Điểm ({self.x}, {self.y}) không nằm trên trục tung")

    def truc_hoanh(self):
        if self.y == 0:
            print(f"Điểm ({self.x}, {self.y}) nằm trên trục hoành")
        else:
            print(f"Điểm ({self.x}, {self.y}) không nằm trên trục hoành")

    def khoang_cach(self):
        d = ((self.x)**2 + (self.y)**2)**0.5
        print(f"Khoảng cách của điểm ({self.x}, {self.y}) tới gốc tọa độ là {round(d, 2)}")
x, y = map(int, input().split())
point = Point(x, y)
point.goc_toa_do()
point.truc_tung()
point.truc_hoanh()
point.khoang_cach()
```

```python
# W8A4
class Calculator:
    def __init__(self, num1, num2):
        self.n1 = float(num1)
        self.n2 = float(num2)

    def add(self):
        print(f"{self.n1} + {self.n2} = {self.n1 + self.n2}")

    def subtract(self):
        print(f"{self.n1} - {self.n2} = {self.n1 - self.n2}")

    def product(self):
        print(f"{self.n1} x {self.n2} = {self.n1 * self.n2}")

    def divide(self):
        print(f"{self.n1} / {self.n2} = {self.n1 / self.n2}")

    def power(self):
        print(f"{self.n1}^{self.n2} = {self.n1 ** self.n2}")

    def mod(self):
        print(f"{self.n1} = {self.n1 % self.n2} (mod {self.n2})")

    def set_numbers(self):
        a, b = map(float, input("Mời nhập 2 phần tử mới: ").split())
        self.n1 = a
        self.n2 = b
        self.menu()

    def menu(self):
        print("1. Cộng")
        print("2. Trừ")
        print("3. Nhân")
        print("4. Chia")
        print("5. Mũ")
        print("6. Đồng dư")
        print("7. Cập nhật 2 số mới")
        print("8. Thoát chương trình")
        while True:
            choice = int(input("Bạn muốn thực hiện phép (Chọn 1 đến 8): "))
            if choice == 1:
                self.add()
                print("1. Quay về menu")
                print("2. Thoát chương trình")
                option = int(input("Mời nhập lựa chọn (1 hoặc 2): "))
                if option == 1:
                    self.menu()
                else:
                    print("Tạm biệt")
                    break
            if choice == 2:
                self.subtract()
                print("1. Quay về menu")
                print("2. Thoát chương trình")
                option = int(input("Mời nhập lựa chọn (1 hoặc 2): "))
                if option == 1:
                    self.menu()
                else:
                    print("Tạm biệt")
                    break
            if choice == 3:
                self.product()
                print("1. Quay về menu")
                print("2. Thoát chương trình")
                option = int(input("Mời nhập lựa chọn (1 hoặc 2): "))
                if option == 1:
                    self.menu()
                else:
                    print("Tạm biệt")
                    break
            if choice == 4:
                self.divide()
                print("1. Quay về menu")
                print("2. Thoát chương trình")
                option = int(input("Mời nhập lựa chọn (1 hoặc 2): "))
                if option == 1:
                    self.menu()
                else:
                    print("Tạm biệt")
                    break
            if choice == 5:
                self.power()
                print("1. Quay về menu")
                print("2. Thoát chương trình")
                option = int(input("Mời nhập lựa chọn (1 hoặc 2): "))
                if option == 1:
                    self.menu()
                else:
                    print("Tạm biệt")
                    break
            if choice == 6:
                self.mod()
                print("1. Quay về menu")
                print("2. Thoát chương trình")
                option = int(input("Mời nhập lựa chọn (1 hoặc 2): "))
                if option == 1:
                    self.menu()
                else:
                    print("Tạm biệt")
                    break
            if choice == 7:
                self.set_numbers()
                print("1. Quay về menu")
                print("2. Thoát chương trình")
                option = int(input("Mời nhập lựa chọn (1 hoặc 2): "))
                if option == 1:
                    self.menu()
                else:
                    print("Tạm biệt")
                    break
            if choice == 8:
                print("Tạm biệt")
                break
            else:
                print("Không có lựa chọn đó, hãy nhập lại")
                continue

x, y = map(str, input("Nhập 2 phần tử cần tính toán: ").split())
cal = Calculator(x, y)
cal.menu()
```

```python
#W8A6
class Rectangle:
    def __init__(self, w, h):
        self.w = w
        self.h = h

    def area(self):
        print(round((self.w * self.h), 2))

    def perimeter(self):
        print(round((2*(self.w + self.h)), 2))

    def scale(self, k):
        self.w *= k
        self.h *= k
        self.area()
        self.perimeter()

w, h, k = map(float, input().split())
rec = Rectangle(w, h)
rec.scale(k)
```

```python
#W8A7
def gcd(a, b):
    if b == 0:
        return a
    else:
        return gcd(b, a%b)
        
class Fraction:
    def __init__(self, num, den):
        if den == 0:
            raise ValueError("Mẫu số không được bằng 0")
        elif den < 0:
            num = -num
            den = -den
        gcd_val = gcd(num, den)
        self.num = num // gcd_val
        self.den = den // gcd_val
        
    def __add__(self, other):
        new_num = self.num * other.den + other.num * self.den
        new_den = self.den * other.den
        return Fraction(new_num, new_den)
        
    def __sub__(self, other):
        new_num = self.num * other.den - other.num * self.den
        new_den = self.den * other.den
        return Fraction(new_num, new_den)
        
    def __mul__(self, other):
        new_num = self.num * other.num
        new_den = self.den * other.den
        return Fraction(new_num, new_den)
        
    def __truediv__(self, other):
        if other.num == 0:
            raise ValueError("Chia cho phân số 0")
        new_num = self.num * other.den
        new_den = self.den * other.num
        return Fraction(new_num, new_den)

    def __str__(self):
        return f"{self.num}/{self.den}"

data = input().split()
a = int(data[0])
b = int(data[1])
op = data[2]
c = int(data[3])
d = int(data[4])

f1 = Fraction(a, b)
f2 = Fraction(c, d)

if op == "+":
    result = f1 + f2
elif op == "-":
    result = f1 - f2
elif op == "*":
    result = f1 * f2
elif op == "/":
    result = f1 / f2
else:
    print("Toán tử không hợp lệ")
    exit()

print(result)
        
```

```python
#W8A8
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = float(balance)

    def deposit(self, x):
        self.balance += float(x)

    def withdraw(self, x):
        if float(x) > self.balance:
            pass
        else:
            self.balance -= float(x)
    
    def __str__(self):
        return f"{self.owner}: {self.balance:.2f}"

#main
data = input().split()
o = data[0]
b = data[1]

account = BankAccount(o, b)

n = int(input())

while n > 0:
    trade = input().split()
    if trade[0] == "DEPOSIT":
        account.deposit(trade[1])
    else:
        account.withdraw(trade[1])
    n -= 1
print(account)
```

```python
#W8A9
class Student:
    def __init__(self, name):
        self.name = name
        self.all_scores = []
        
    def add_score(self, subject, score):
        self.all_scores.append(float(score))
        
    def avg(self):
        avg = sum(self.all_scores)/len(self.all_scores)
        return avg
        
    def rank(self):
        if self.avg() >= 8:
            return "Excellent"
        elif self.avg() >= 6.5:
            return "Good"
        elif self.avg() >= 5:
            return "Average"
        else:
            return "Poor"
    def __str__(self):
        self.avg()
        self.rank()
        return f"{self.name} {self.avg():.2f} {self.rank()}"

name = Student(input())
n = int(input())
while n > 0:
    data = input().split()
    name.add_score(data[0], data[1])
    n -= 1
print(name)
```

```python
#W8A10
class Book:
    def __init__(self, title, author, year):
        self.title = title
        self.author = author
        self.year = int(year)

class Library():
    def __init__(self):
        self.book_list = []

    def add(self, book):
        self.book_list.append(book)

    def count_by_author(self, name):
        count = 0
        for i in self.book_list:
            if i.author == name:
                count += 1
        return count

    def find_by_year(self, y):
        count = 0
        for i in self.book_list:
            if i.year == int(y):
                count += 1
        return count

n = int(input())
big_data = Library()
while n > 0:
    x = input().split()
    if x[0] == "ADD":
        parts = x[1].split(";")
        data = Book(parts[0], parts[1], parts[2])
        big_data.add(data)
    elif x[0] == "COUNT":
        print(big_data.count_by_author(x[1]))
    else:
        print(big_data.find_by_year(x[1]))
    n -= 1
```
