```python
#W6A1
def list_rut_gon(l):
    result = []
    seen = set()
    for i in l:
        if i not in seen:
            result.append(i)
            seen.add(i)
    return result
l = list(map(int, input().split()))
print(list_rut_gon(l))
```

```python
#W6A2
def tong_tich_luy(l):
    for i in range(len(l)):
        if i == 0:
            l[i] = l[i]
        else:
            l[i] += l[i-1]
    return l
l = list(map(int, input().split()))
print(tong_tich_luy(l))
```

```python
#W6A3
def xoay(l, k):
    a = 0
    for i in range(k):
        a = l[0]
        for j in range(len(l)):
            if j == len(l) - 1:
                l[j] = a
            else:
                l[j] = l[j+1]
    return tuple(l)
l = list(map(int, input().split()))
k = int(input())
print(xoay(l, k))
```

```python
#W6A5
def dem_so(l):
    result = {}
    po = 0
    ne = 0
    zeros = 0
    for i in range(len(l)):
        if l[i] > 0:
            po += 1
        elif l[i] < 0:
            ne += 1
        else:
            zeros += 1
    result['positives'] = po
    result['negatives'] = ne
    result['zeros'] = zeros
    return result
l = list(map(int, input().split()))
print(dem_so(l))
```

```python
#W6A9
def gop_dict(l1, l2):
    result = {}
    for key in l1:
        if key in l2:
            result[key] = l1[key] + l2[key] 
        else:
            result[key] = l1[key]           
    for key in l2:
        if key not in result:
            result[key] = l2[key]
    return {key: result[key] for key in sorted(result)}
l1 = eval(input())
l2 = eval(input())
print(gop_dict(l1, l2))
```

```python
#W6A10:
def tong(l, k):
    result = set()
    for i in range(len(l)):
        for j in range(i + 1, len(l)):
            if l[i] + l[j] == k:
                result.add((l[i], l[j]))
    return sorted(list(result))
l = list(map(int, input().split()))
k = int(input())
print(tong(l, k))
```

```python
#W6A19
def position_max_of_list(l):
    max_result = l[0]
    index_max_result = 0
    for i in range(len(l)):
        if l[i] >= max_result:
            max_result = l[i]
            index_max_result = i + 1
    return index_max_result
l = list(map(int, input().split()))
print(position_max_of_list(l))
```

```python
#W6A20
def find_value(l, k):
    result = -1
    for i in range(len(l) - 1, -1, -1):
        if l[i] == k:
            result = i + 1
    return result
l = list(map(int, input().split()))
k = int(input())
print(find_value(l, k))
```
