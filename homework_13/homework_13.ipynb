```python
#W11A1
def reverse(path: str) -> int:
    try:
        with open(path, 'r') as file:
            content = file.read().strip()
        reversed_num = int(content[::-1])
        return reversed_num
    except FileNotFoundError:
        print("Lỗi: Không tìm thấy tệp tin.")
    except ValueError:
        print("Lỗi: Nội dung tệp không phải là số nguyên.")
reverse("W11A1.txt")
```

```python
#W11A4
import string

def clean_string(text: str) -> str:
    text = text.lower()
    return "".join(char for char in text if char.isalnum())

def areAnagrams(path1: str, path2: str) -> bool:
    try:
        with open(path1, 'r', encoding='utf-8') as f1, \
             open(path2, 'r', encoding='utf-8') as f2:
            str1 = f1.read()
            str2 = f2.read()
            
        clean1 = clean_string(str1)
        clean2 = clean_string(str2)
        
        return sorted(clean1) == sorted(clean2)
        
    except FileNotFoundError:
        print("Lỗi: Một trong hai tệp không tồn tại.")
        return False
    except ValueError:
        print("Lỗi: Định dạng tệp tin không hợp lệ.")
        return False
areAnagrams("W11A4_1.txt", "W11A4_2.txt")
```

```python
#W11A5
def count(path1: str, character: str) -> int:
    try:
        with open(path1, 'r', encoding='utf-8') as file:
            s = file.read().strip()

        counts = 0
        for i in range(len(s)):
            if s[i] == character:
                counts += 1
        return counts
        
    except FileNotFoundError:
        print("Lỗi: Không tìm thấy tệp tin.")
        return False
    except ValueError:
        print("Lỗi: Định dạng tệp tin không hợp lệ.")
        return False
count("W11A5.txt", "e")
```

```python
#W11A8
def xac_dinh_tu_fibonacci():
    try:
        n = int(input("n = "))
        fib = []
        
        if n >= 0:
            fib.append("a")
        if n >= 1:
            fib.append("b")
            
        for i in range(2, n + 1):
            tu_moi = fib[i-1] + fib[i-2]
            fib.append(tu_moi)

        with open("W11A8.txt", "w", encoding="utf-8") as f:
            for i in range(len(fib)):
                line = f"f({i}) = {fib[i]}\n"
                f.write(line)
                print(line.strip())

    except ValueError:
        print("Lỗi: n phải là một số nguyên.")
    except Exception as e:
        print(f"Đã xảy ra lỗi: {e}")

xac_dinh_tu_fibonacci()
```

```python
#W11A10
def getMoney(path: str) -> list[int]:
    try:
        with open(path, "r", encoding="utf-8") as f:
            list_money = f.read().strip().split()
        for i in range(len(list_money)):
            list_money[i] = int(list_money[i])
        return list_money
    except FileNotFoundError:
        print("Lỗi: Không tìm thấy tệp tin")
def rob(houses: list[int]) -> int:
    sum_1 = 0
    sum_2 = 0
    for i in range(len(houses)):
        if i % 2 == 0:
            sum_1 += houses[i]
        else:
            sum_2 += houses[i]
    return max(sum_1, sum_2)
print(getMoney("W11A10.txt"))
print(rob(getMoney("W11A10.txt")))
```

```python
#W11A12
def maximumProduct(path: str) -> int:
    try:
        with open(path, 'r', encoding='utf-8') as f:
            n = int(f.readline().strip())
            data = f.readline().split()
            nums = [int(x) for x in data]
            
            if len(nums) < 3:
                raise ValueError("Mảng phải có ít nhất 3 phần tử")

            max1 = max2 = max3 = float('-inf')
            min1 = min2 = float('inf')
            
            for x in nums:
                if x > max1:
                    max1, max2, max3 = x, max1, max2
                elif x > max2:
                    max2, max3 = x, max2
                elif x > max3:
                    max3 = x
                
                if x < min1:
                    min1, min2 = x, min1
                elif x < min2:
                    min2 = x
            
            return max(max1 * max2 * max3, min1 * min2 * max1)

    except Exception as e:
        print(f"Lỗi: {e}")
        return 0

maximumProduct("W11A12.txt")
```

```python
#W11A13
def moveZeroes(fileName: str) -> list:
    try:
        with open(fileName, "r", encoding="utf-8") as f:
            n = int(f.readline().strip())
            l = f.readline().strip().split()

            arr = [int(x) for x in l]
            
            if len(arr) != n:
                raise ValueError(f"Mảng khác {n} phần tử")
            zero_list = []
            not_zero_list = []
            for i in range(len(arr)):
                if arr[i] == 0:
                    zero_list.append(arr[i])
                else:
                    not_zero_list.append(arr[i])
            result = not_zero_list + zero_list
            return result
    except Exception as e:
        print(f"Lỗi: {e}")

moveZeroes("W11A13.txt")
```

```python
#W11A14
import csv

def averageTime(path: str):
    total_time = 0.00
    lap_count = 0
    
    try:
        with open(path, mode='r', encoding='utf-8') as file:
            reader = csv.DictReader(file)
            
            for row in reader:
                total_time += float(row['Time'])
                lap_count += 1
        
        if lap_count == 0:
            return 0.00
        average = total_time / lap_count
        return float(f"{average:.2f}")
        
    except Exception as e:
        return f"Lỗi: {e}"

averageTime("W11A14.csv")
```

```python
#W11A15
import csv

def findMovies(path: str) -> list[str]:
    suitable_movies = []
    
    try:
        with open(path, mode='r', encoding='utf-8') as file:
            reader = csv.DictReader(file)
            
            for row in reader:
                movie_name = row['movie'].strip()
                description = row['description'].strip().lower()
                rating = float(row['rating'])
                
                if rating >= 8.0 and description != "boring":
                    suitable_movies.append(movie_name)
        
        if not suitable_movies:
            return "All are boring!"
            
        return suitable_movies

    except Exception as e:
        return f"Lỗi: {e}"
findMovies("W11A15.csv")
```

```python
#W11A16
import math

def calculate_perimeters(path: str):
    try:
        with open(path, 'r', encoding='utf-8') as file:
            lines = [line.strip() for line in file if line.strip()]
            
        for i in range(0, len(lines), 2):
            type_line = lines[i]      
            data_line = lines[i+1]    
            
            shape_type = type_line.split(":")[1].strip()
            
            perimeter = 0.0
            
            if shape_type == "SQUARE":
                side = float(data_line.split(":")[1].strip())
                perimeter = 4 * side
                
            elif shape_type == "RECTANGLE":
                parts = data_line.split()
                width = float(parts[1])
                height = float(parts[3])
                perimeter = 2 * (width + height)
                
            elif shape_type == "CIRCLE":
                radius = float(data_line.split(":")[1].strip())
                perimeter = 2 * math.pi * radius
            
            print(f"{perimeter:.2f}")
            

    except Exception as e:
        print(f"Lỗi: {e}")

calculate_perimeters("W11A16.txt")
```

```python
#W11A17
def productExceptSelf(path: str) -> list[int]:
    try:
        with open(path, 'r', encoding='utf-8') as file:
            n = int(file.readline().strip())
            content = file.read()
            nums = [int(x) for x in content.split()]
        
        n = len(nums)
        if n <= 1:
            return []

        res = [1] * n
        
        left_product = 1
        for i in range(n):
            res[i] = left_product
            left_product *= nums[i]
            
        right_product = 1
        for i in range(n - 1, -1, -1):
            res[i] *= right_product
            right_product *= nums[i]
        print(*res)

    except Exception as e:
        print(f"Lỗi: {e}")

productExceptSelf("W11A17.txt")
```

```python
#W11A18
import math
import re

def estimatedTime(path: str):
    try:
        src_x = src_y = des_x = des_y = velocity = None
        
        with open(path, 'r', encoding='utf-8') as file:
            content = file.read()
            
            # Sử dụng Regex (biểu thức chính quy) để tìm số sau các từ khóa
            src_match = re.search(r"SRC\s+X:\s*([\d.-]+)\s+Y:\s*([\d.-]+)", content)
            des_match = re.search(r"DES\s+X:\s*([\d.-]+)\s+Y:\s*([\d.-]+)", content)
            vel_match = re.search(r"VELOCITY:\s*([\d.-]+)", content)
            
            if src_match:
                src_x, src_y = map(float, src_match.groups())
            if des_match:
                des_x, des_y = map(float, des_match.groups())
            if vel_match:
                velocity = float(vel_match.group(1))

        if None in [src_x, src_y, des_x, des_y, velocity]:
            return "Lỗi: Tệp tin thiếu thông tin hoặc sai định dạng."

        # Tính khoảng cách và thời gian
        distance = math.sqrt((des_x - src_x)**2 + (des_y - src_y)**2)
        time_seconds = distance / velocity
        
        return round(time_seconds, 2)

    except Exception as e:
        return f"Lỗi: {e}"

estimatedTime("W11A18.txt")
```

```python
#W11A19
import struct

def read_binary_doubles(path: str):
    
    try:
        with open(path, 'rb') as file:
            content = file.read()
            
            double_size = 8
            
            n = len(content) // double_size
            
            # Giải mã (unpack) dữ liệu nhị phân thành danh sách các số thực
            # Định dạng 'd' là double, n là số lượng phần tử
            format_string = f'{n}d'
            numbers = struct.unpack(format_string, content[:n * double_size])
            

            for num in numbers:
                print(f"{num:.5f}")
                
    except Exception as e:
        print(f"Lỗi: {e}")

read_binary_doubles("W11A19.bin")
```

```python
#W11A20
import math

def imageSmoother(path: str):
    try:
        with open(path, 'r', encoding='utf-8') as file:
            M = int(file.readline().strip())
            
            img = []
            for _ in range(M):
                row = list(map(int, file.readline().split()))
                img.append(row)
        
        result = [[0] * M for _ in range(M)]
        
        for r in range(M):
            for c in range(M):
                total_sum = 0
                count = 0
                
                # Duyệt cửa sổ 3x3 xung quanh pixel (r, c)
                # i, j chạy từ -1 đến 1 để quét các ô lân cận
                for i in range(-1, 2):
                    for j in range(-1, 2):
                        nr, nc = r + i, c + j
                        
                        # Kiểm tra xem pixel lân cận có nằm trong phạm vi ma trận không
                        if 0 <= nr < M and 0 <= nc < M:
                            total_sum += img[nr][nc]
                            count += 1
                
                # Tính trung bình và làm tròn xuống (floor)
                result[r][c] = total_sum // count
        
        # In ma trận sau khi làm mịn
        for row in result:
            print(*(row))
            
    except Exception as e:
        print(f"Lỗi: {e}")

imageSmoother("W11A20.txt")
```
