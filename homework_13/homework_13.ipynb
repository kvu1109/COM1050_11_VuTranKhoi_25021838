```python
#W11A1
def reverse(path: str) -> int:
    try:
        with open(path, 'r') as file:
            content = file.read().strip()
        reversed_num = int(content[::-1])
        return reversed_num
    except FileNotFoundError:
        print("Lỗi: Không tìm thấy tệp tin.")
    except ValueError:
        print("Lỗi: Nội dung tệp không phải là số nguyên.")
reverse("W11A1.txt")
```

```python
#W11A2
def canWinNim(path: str) -> bool:
    try:
        # 1. Mở và đọc số lượng đá từ tệp
        with open(path, 'r') as file:
            content = file.read().strip()
            n = int(content)
        return n % 4 != 0  
    except FileNotFoundError:
        print("Lỗi: Không tìm thấy tệp tin.")
        return False
    except ValueError:
        print("Lỗi: Nội dung tệp không phải là số nguyên.")
        return False
canWinNim("W11A2.txt")     
```

```python
#W11A3
def housesOfHogwarts(path: str):
    try:
        with open(path, 'r', encoding='utf-8') as file:
            line = file.readline().strip()
            if not line:
                return
            
            n = int(line)
            
            for i in range(n):
                name = file.readline().strip()
                if name:
                    print(name)
                    
    except FileNotFoundError:
        print("Lỗi: Không tìm thấy tệp tin.")
    except ValueError:
        print("Lỗi: Định dạng tệp tin không hợp lệ.")
housesOfHogwarts("W11A3.txt")
```

```python
#W11A4
import string

def clean_string(text: str) -> str:
    text = text.lower()
    return "".join(char for char in text if char.isalnum())

def areAnagrams(path1: str, path2: str) -> bool:
    try:
        with open(path1, 'r', encoding='utf-8') as f1, \
             open(path2, 'r', encoding='utf-8') as f2:
            str1 = f1.read()
            str2 = f2.read()
            
        clean1 = clean_string(str1)
        clean2 = clean_string(str2)
        
        return sorted(clean1) == sorted(clean2)
        
    except FileNotFoundError:
        print("Lỗi: Một trong hai tệp không tồn tại.")
        return False
    except ValueError:
        print("Lỗi: Định dạng tệp tin không hợp lệ.")
        return False
areAnagrams("W11A4_1.txt", "W11A4_2.txt")
```

```python
#W11A5
def count(path1: str, character: str) -> int:
    try:
        with open(path1, 'r', encoding='utf-8') as file:
            s = file.read().strip()

        counts = 0
        for i in range(len(s)):
            if s[i] == character:
                counts += 1
        return counts
        
    except FileNotFoundError:
        print("Lỗi: Không tìm thấy tệp tin.")
        return False
    except ValueError:
        print("Lỗi: Định dạng tệp tin không hợp lệ.")
        return False
count("W11A5.txt", "e")
```
