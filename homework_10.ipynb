```python
#W9A1
def merge_list(l1, l2):
    if not l1:
        return l2
    elif not l2:
        return l1
    elif l1[0] < l2[0]:               
        return [l1[0]] + merge_list(l1[1:], l2)
    else:
        return [l2[0]] + merge_list(l1, l2[1:])
l1 = list(map(int, input().split()))
l2 = list(map(int, input().split()))

print(*merge_list(l1, l2))
```

```python
#W9A2
def find_kth_largest(matrix, k):
    all_numbers = []
    for row in matrix:
        all_numbers.extend(row)
    all_numbers.sort()
    return all_numbers[k-1]
  
n, m = map(int, input().split())
matrix = [list(map(int, input().split())) for rows in range(n)]
k = int(input())
print(find_kth_largest(matrix, k)) 
```

```python
#W9A3
def find_min_idx_after(s_list, start_idx):
    min_idx = start_idx
    for j in range(len(s_list) - 1, start_idx, -1):
        if s_list[j] < s_list[min_idx]:
            min_idx = j
    return min_idx
def swap_to_get_smallest(n):
    s = list(n)
    for i in range(len(s)):
        m_idx = find_min_idx_after(s, i)
        if s[m_idx] < s[i]:
            s[i], s[m_idx] = s[m_idx], s[i]
            return "".join(s)
    
    s[-1], s[-2] = s[-2], s[-1]
    return "".join(s)

n = input()
print(int(swap_to_get_smallest(n)))
```

```python
#W9A4
def is_transform(s1, s2):
    n, m = len(s1), len(s2)
    dp = [[False] * (m + 1) for rows in range(n + 1)]
    dp[0][0] = True
    
    for i in range(1, n + 1):
        for j in range(m + 1):
            char_s1 = s1[i-1]
            if dp[i-1][j]:
                if char_s1.islower():
                    dp[i][j] = True
            if j > 0 and dp[i-1][j-1]:
                char_s2 = s2[j-1]
                if char_s1.upper() == char_s2:
                    dp[i][j] = True
                    
    return "YES" if dp[n][m] else "NO" 
q = int(input())
for steps in range(q):
    s1 = str(input())
    s2 = str(input())
    print(is_transform(s1, s2))
```

```python
#W9A5
def is_sublist(l1, l2):
    if len(l1) == 0:
        return 'YES'
    else:
        for i in range(len(l2) - len(l1) + 1):
            if l2[i : i + len(l1)] == l1:
                return 'YES'
    return 'NO'
l1 = list(map(int, input().split()))
l2 = list(map(int, input().split()))
print(is_sublist(l1, l2))
```

```python
#W9A6
def hoan_vi_tiep_theo(l):
    for i in range(len(l)-2,-1,-1):
        if l[i] < l[i+1]:
            for j in range(len(l)-1, i, -1):
                if l[j] > l[i]:
                    l[i], l[j] = l[j], l[i]
                    l[i+1:len(l)] = l[len(l)-1:i:-1]
                    return l
    return l[::-1]
n = int(input())
l = list(map(int, input().split()))
print(*hoan_vi_tiep_theo(l))
```

```python
#W9A7
def canPlaceFlowers(flowerbed, k):
    placeable_slots = 0
    if k == 0:
        return True
    elif len(flowerbed) == 1 and flowerbed[0] == 0:
        placeable_slots += 1
    elif len(flowerbed) == 1 and flowerbed[0] == 1:
        return False
    else:
        for i in range(len(flowerbed)):
            if i == 0 and flowerbed[i+1] == 0 and flowerbed[i] == 0 or i == len(flowerbed)-1 and flowerbed[i-1] == 0 and flowerbed[i] == 0:
                flowerbed[i] = 1
                placeable_slots += 1
            elif flowerbed[i] == 0 and flowerbed[i-1] == 0 and flowerbed[i+1] == 0:
                flowerbed[i] = 1
                placeable_slots += 1
    if placeable_slots >= k:
        return True
    else:
        return False
f = list(map(int, input().split()))
k = int(input())
print(canPlaceFlowers(f, k))
```

```python
#W9A8
def location_after_moves(goc, move):
    if move == "R":
        goc[1] += 1
    elif move == "L":
        goc[1] -= 1
    elif move == "U":
        goc[0] += 1
    elif move == "D":
        goc[0] -= 1
    return goc

def is_path_crossing(moves):
    loc = [0, 0]
    location_passed = {tuple(loc)}
    for i in moves:
        loc = location_after_moves(loc, i) 
        current_loc = tuple(loc)
        if current_loc in location_passed:
            return True
        else:
            location_passed.add(current_loc)
    return False

l = str(input())
print(is_path_crossing(l))
    
```

```python
#W9A9
def Median(nums):
    a, b, c, d, e = nums
    if a > b: 
        a, b = b, a 
    if c > d: 
        c, d = d, c 
    if a > c:             
        a, c = c, a
        b, d = d, b
    if e > c:             
        if e > d:        
            if b > d:
                return b
            else:
                return d 
        else:
            if b > e:
                return b
            else:
                return e
    else:
        if b > c:        
            return c
        else:
            return b
l = list(map(int, input().split()))
print(Median(l))
```

```python
#W9A10
def find_insert_index(arr, value):
    left, right = 0, len(arr) - 1
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == value:
            return mid
        elif arr[mid] < value:
            left = mid + 1
        else:
            right = mid - 1
    return left
    
def findRadius(houses, heaters):
    heaters.sort()
    max_radius = 0
    
    for house in houses:
        i = find_insert_index(heaters, house)
        if i < len(heaters):
            dist_right = heaters[i] - house
        else:
            dist_right = float('inf')
        if i > 0:
            dist_left = house - heaters[i-1]
        else:
            dist_left = float('inf')
        current_min_dist = min(dist_left, dist_right)
        max_radius = max(max_radius, current_min_dist)
        
    return max_radius
a = list(map(int, input().split()))
b = list(map(int, input().split()))
print(findRadius(a, b))
```
