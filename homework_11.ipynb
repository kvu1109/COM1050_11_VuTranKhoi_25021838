```python
#W8A3
def tong_binh_phuong_chu_so(n):
    s = str(n)
    sum = 0
    for i in range(len(s)):
        sum += int(s[i])**2
    return sum
def ishappynumber(n):
    seen = set()
    current = n
    while current != 1 and current not in seen:
        seen.add(current)
        current = tong_binh_phuong_chu_so(current)
    if current == 1:
        return "YES"
    else:
        return "NO"
try:
    a = int(input())
    print(ishappynumber(a))
except ValueError:
    print(f"invalid literal for int() with base 10: '{a}'")
```

```python
#W10A10
try:
    n = int(input())
    array = list(map(int, input().split()))
    phan_tu_da_co = set()
    sum_mang = 0
    if len(array) != n:
        print(f"So phan tu trong chuoi khac {n}")
    else:
        for i in range(len(array)):
            if array[i] <= 0:
                print("Mang phai gom cac so nguyen duong")
                break
            elif array[i] in phan_tu_da_co:
                break
            else:
                phan_tu_da_co.add(array[i])
                sum_mang += array[i]
        if sum(array) != sum_mang:
            print("Mang khong hop le")
        else:
            print(sum_mang)
except ValueError:
    print("Chuoi khong hop le")
```
