```python
#Phân loại tam giác
a, b, c = map(int, input().split())
if a + b <= c or b + c <= a or c + a <= b or a <= 0 or b <= 0 or c <= 0:
  print("Không phải tam giác")
elif a == b == c:
  print("Tam giác đều")
elif a == b or b == c or a == c:
  print("Tam giác cân")
else:
  print("Tam giác thường")
```

```python
#Tính tiền điện
n = int(input())
if n <= 50:
  print(n*1500)
elif n <= 100:
  print(50*1500 + (n-50)*2000)
else:
  print(50*1500 + 50*2000 + (n-100)*3000)
```

```python
#Rounding
x = float(input())
if x // 1 == x:
  print(int(x), int(x), int(x))
elif x < 0:
  round_down = int(x // 1)
  round_up = int(round_down + 1)
  if round_up - x < x - round_down:
    print(round_up, round_down, round_up)
  else:
    print(round_up, round_down, round_down)
else:
  round_up = int(x // 1 + 1)
  round_down = int(x // 1)
  if round_up - x < x - round_down:
    print(round_up, round_down, round_up)
  else:
    print(round_up, round_down, round_down)
```
