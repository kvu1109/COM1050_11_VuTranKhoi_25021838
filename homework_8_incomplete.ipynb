```python
#W8A1
class Cylinder:
    def __init__(self, radius, height):
        self.radius = radius
        self.height = height

    def calculate_cylinder(self):
        r = self.radius
        h = self.height
        pi = 3.14
        
        a = 2*pi*r*h + 2*pi*(r**2)
        v = pi*(r**2)*h
        print(round(a, 2), round(v, 2))
r, h = map(float, input().split())
c = Cylinder(r, h)
c.calculate_cylinder()
```

```python
#W8A3
class Point:
    def __init__(self, x_val, y_val):
        self.x = x_val
        self.y = y_val

    def goc_toa_do(self):
        if self.x == 0 and self.y == 0:
            print(f"Điểm ({self.x}, {self.y}) là gốc tọa độ")
        else:
            print(f"Điểm ({self.x}, {self.y}) không là gốc tọa độ")

    def truc_tung(self):
        if self.x == 0:
            print(f"Điểm ({self.x}, {self.y}) nằm trên trục tung")
        else:
            print(f"Điểm ({self.x}, {self.y}) không nằm trên trục tung")

    def truc_hoanh(self):
        if self.y == 0:
            print(f"Điểm ({self.x}, {self.y}) nằm trên trục hoành")
        else:
            print(f"Điểm ({self.x}, {self.y}) không nằm trên trục hoành")

    def khoang_cach(self):
        d = ((self.x)**2 + (self.y)**2)**0.5
        print(f"Khoảng cách của điểm ({self.x}, {self.y}) tới gốc tọa độ là {round(d, 2)}")
x, y = map(int, input().split())
point = Point(x, y)
point.goc_toa_do()
point.truc_tung()
point.truc_hoanh()
point.khoang_cach()
```

```python
# W8A4
class Calculator:
    def __init__(self, num1, num2):
        self.n1 = float(num1)
        self.n2 = float(num2)

    def add(self):
        print(f"{self.n1} + {self.n2} = {self.n1 + self.n2}")

    def subtract(self):
        print(f"{self.n1} - {self.n2} = {self.n1 - self.n2}")

    def product(self):
        print(f"{self.n1} x {self.n2} = {self.n1 * self.n2}")

    def divide(self):
        print(f"{self.n1} / {self.n2} = {self.n1 / self.n2}")

    def power(self):
        print(f"{self.n1}^{self.n2} = {self.n1 ** self.n2}")

    def mod(self):
        print(f"{self.n1} = {self.n1 % self.n2} (mod {self.n2})")

    def set_numbers(self):
        a, b = map(float, input("Mời nhập 2 phần tử mới: ").split())
        self.n1 = a
        self.n2 = b
        self.menu()

    def menu(self):
        print("1. Cộng")
        print("2. Trừ")
        print("3. Nhân")
        print("4. Chia")
        print("5. Mũ")
        print("6. Đồng dư")
        print("7. Cập nhật 2 số mới")
        print("8. Thoát chương trình")
        while True:
            choice = int(input("Bạn muốn thực hiện phép (Chọn 1 đến 8): "))
            if choice == 1:
                self.add()
                print("1. Quay về menu")
                print("2. Thoát chương trình")
                option = int(input("Mời nhập lựa chọn (1 hoặc 2): "))
                if option == 1:
                    self.menu()
                else:
                    print("Tạm biệt")
                    break
            if choice == 2:
                self.subtract()
                print("1. Quay về menu")
                print("2. Thoát chương trình")
                option = int(input("Mời nhập lựa chọn (1 hoặc 2): "))
                if option == 1:
                    self.menu()
                else:
                    print("Tạm biệt")
                    break
            if choice == 3:
                self.product()
                print("1. Quay về menu")
                print("2. Thoát chương trình")
                option = int(input("Mời nhập lựa chọn (1 hoặc 2): "))
                if option == 1:
                    self.menu()
                else:
                    print("Tạm biệt")
                    break
            if choice == 4:
                self.divide()
                print("1. Quay về menu")
                print("2. Thoát chương trình")
                option = int(input("Mời nhập lựa chọn (1 hoặc 2): "))
                if option == 1:
                    self.menu()
                else:
                    print("Tạm biệt")
                    break
            if choice == 5:
                self.power()
                print("1. Quay về menu")
                print("2. Thoát chương trình")
                option = int(input("Mời nhập lựa chọn (1 hoặc 2): "))
                if option == 1:
                    self.menu()
                else:
                    print("Tạm biệt")
                    break
            if choice == 6:
                self.mod()
                print("1. Quay về menu")
                print("2. Thoát chương trình")
                option = int(input("Mời nhập lựa chọn (1 hoặc 2): "))
                if option == 1:
                    self.menu()
                else:
                    print("Tạm biệt")
                    break
            if choice == 7:
                self.set_numbers()
                print("1. Quay về menu")
                print("2. Thoát chương trình")
                option = int(input("Mời nhập lựa chọn (1 hoặc 2): "))
                if option == 1:
                    self.menu()
                else:
                    print("Tạm biệt")
                    break
            if choice == 8:
                print("Tạm biệt")
                break
            else:
                print("Không có lựa chọn đó, hãy nhập lại")
                continue

x, y = map(str, input("Nhập 2 phần tử cần tính toán: ").split())
cal = Calculator(x, y)
cal.menu()
```

```python
#W8A6
class Rectangle:
    def __init__(self, w, h):
        self.w = w
        self.h = h

    def area(self):
        print(round((self.w * self.h), 2))

    def perimeter(self):
        print(round((2*(self.w + self.h)), 2))

    def scale(self, k):
        self.w *= k
        self.h *= k
        self.area()
        self.perimeter()

w, h, k = map(float, input().split())
rec = Rectangle(w, h)
rec.scale(k)
```
