```python
#W7A1
def binary_search(arr, value):
    left, right = 0, len(arr) - 1
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == value:
            return mid
        elif arr[mid] < value:
            left = mid + 1
        else:
            right = mid - 1
    return -1
l = list(map(int, input().split()))
t = int(input())
print(binary_search(l, t))
```

```python
#W7A3
def bubble_sort(arr):
    for i in range(len(arr)):
        swap = False
        for j in range(len(arr) - i - 1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                swap = True
        if not swap:
            break
    return arr

def selection_sort(arr):
    for i in range(len(arr)):
        min_idx = i
        for j in range(i+1, len(arr)):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr

l = list(map(int, input().split()))
print(bubble_sort(l))
print(selection_sort(l))       
```

```python
#W7A4
def most_appear(arr):
    most_appear_count = 0
    most_appear_value = 0
    for i in range(len(arr) - 1, -1, -1):
        count = 1
        for j in range(i - 1, -1, -1):
            if arr[i] == arr[j]:
                count += 1
        if count >= most_appear_count:
            most_appear_count = count
            most_appear_value = arr[i]
    return f"{most_appear_value} xuat hien nhieu nhat, som nhat, {most_appear_count} lan"
l = list(map(int, input().split()))
print(most_appear(l))
```

```python
#W7A5
def dem_cap_bang_tong(arr, x):
    count = 0
    for i in range(len(arr)):
        for j in range(i+1, len(arr)):
            if x - arr[i] == arr[j]:
                count += 1
    return count
l = list(map(int, input().split()))
x = int(input())
print(dem_cap_bang_tong(l, x))
```

```python
#W7A7
def longest_increasing_subsequence_length(arr):
    dp = [1] * len(arr)
    for i in range(1, len(arr)):
        for j in range(i):
            if arr[j] < arr[i]:
                dp[i] = max(dp[i], dp[j] + 1)
    return max(dp)

arr = eval(input())
print(longest_increasing_subsequence_length(arr))
```

```python
#W7A8
def min_range(interval, queries):
    dic = {}
    for i in queries:
        dic[i] = []
        for j in range(len(interval)):
            if i >= interval[j][0] and i <= interval[j][1]:
                dic[i].append(interval[j][1] - interval[j][0] + 1)
    result = []
    for i in dic:
        if len(dic[i]) == 0:
            dic[i].append(-1)
        result.append(min(dic[i]))
    return result
i = eval(input())
q = eval(input())
print(min_range(i, q))
```

```python
#W7A9
def count_smaller_right(arr):
    result = [0]*len(arr)
    for i in range(len(arr)):
        for j in range(i+1, len(arr)):
            if arr[j] < arr[i]:
                result[i] += 1
    return result
arr = list(map(int, input().split(',')))
print(count_smaller_right(arr))
```

```python
#W7A10
import bisect

def max_subarray_mod(nums, m):
    prefix_sum = 0
    max_mod = 0
    
    seen_prefixes = [0]
    
    for x in nums:
        prefix_sum = (prefix_sum + x) % m
        max_mod = max(max_mod, prefix_sum)

        idx = bisect.bisect_right(seen_prefixes, prefix_sum)
        
        if idx < len(seen_prefixes):
            max_mod = max(max_mod, (prefix_sum - seen_prefixes[idx] + m) % m)
        
        bisect.insort(seen_prefixes, prefix_sum)
        
    return max_mod

l = eval(input())
n = int(input())
print(max_subarray_mod(l, n))
```

```python
#W7A13
def find_longest_harmonious_subsequence(l):
    counts = {}
    for i in l:
        counts[i] = counts.get(i, 0) + 1
    
    max_length = 0
    
    for i in counts:
        if i + 1 in counts:
            current_length = counts[i] + counts[i + 1]
            max_length = max(max_length, current_length)
            
    return max_length

arr = list(map(int, input().split()))
print(find_longest_harmonious_subsequence(arr))
```

```python
#W7A14
def find_7(l):
    result = []
    for i in range(len(l)-1,-1,-1):
        if l[i] == 7:
            result.append(i)
    return result
    
n = int(input())
l = list(map(int, input().split()))
if len(l) != n:
    print(f"Cần nhập {n} phần tử")
else:
    if len(find_7(l)) == 0:
        print("Not found")
    else:
        print(*find_7(l))
```
