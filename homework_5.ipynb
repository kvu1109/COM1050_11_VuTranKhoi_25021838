```python
#W5A1
def max_of_two(a, b):
    if a >= b:
        return a
    else:
        return b
a, b = map(int, input().split())
print(max_of_two(a, b))
```

```python
#W5A2
def swap(a, b):
    c = a
    a = b
    b = c
    return a, b
a, b = map(int, input().split())
print(*swap(a, b))
```

```python
#W5A3
def is_prime(n):
    if n <= 1:
        return False
    else:
        for i in range(2, int(n**0.5) + 1):
            if n % i == 0:
                return False
        return True
n = int(input())
print(is_prime(n))
```

```python
#W5A4
def perfect_number(n):
    sum = 0
    for i in range(1, n):
        if n % i == 0:
            sum += i
    if sum == n:
        return True
    else:
        return False
n = int(input())
print(perfect_number(n))
```

```python
#W5A5
def find_number(lst, k):
    for i in range(len(lst)):
        if lst[i] == k:
            return i + 1
    return -1
lst = list(map(int, input().split()))
k = int(input())
print(find_number(lst, k))
```

```python
#W5A6
def factorial(n):
    fac = 1
    for i in range(1, n+1):
        fac *= i
    return fac
n = int(input())
print(factorial(n))
```

```python
#W5A7
def caculator(num1, operat, num2):
    num1 = float(num1)
    num2 = float(num2)
    if operat == "+":
        return num1 + num2
    elif operat == "-":
        return num1 - num2
    elif operat == "/":
        return num1 / num2
    elif operat == "*":
        return num1 * num2
a, b, c = map(str, input().split())
print(round(caculator(a, b, c), 2))
```

```python
#W5A8 (Incomplete)
def hamming_range(x, y):
    a = str(bin(x)[2:])
    b = str(bin(y)[2:])
    count = 0
    if len(a) <= len(b):
        for i in range(len(a)):
            if a[i] != b[i]:
                count += 1
    else:
        for i in range(len(b)):
            if a[i] != b[i]:
                count += 1
    return count + abs(len(a) - len(b))
a, b = map(int, input().split())
print(hamming_range(a, b))
```

```python
#W5A9
def digits_sum(n):
    sum = 0
    while n >= 1:
        sum += n % 10
        n //= 10
    return sum
n = int(input())
print(digits_sum(n))
```

```python
#W5A10
def dang_cau(first_word, second_word):
    #Tạo từ điển rỗng để tra ký tự từ fw sang sw
    map_fw_to_sw = {}
    #Bảng đánh dấu ký tự của sw đã dùng chưa
    used_sw = set()
    for i in range(len(first_word)):
        char_fw = first_word[i]
        char_sw = second_word[i]
        #Kiểm tra xem ký tự của từ đầu tiên đã xuất hiện trong từ điển chưa
        if char_fw in map_fw_to_sw:
            #Kiểm tra xem ký tự của từ đầu tiên đấy có tra ra ký tự của từ thứ hai không, nếu không thì trả False
            if map_fw_to_sw.get(char_fw) != char_sw:
                return False
        #Nếu ký tự từ đầu tiên chưa có trong từ điển
        else:
            #Kiểm tra ký tự từ thứ hai đã bị sử dụng chưa, nếu có thì trả False
            if char_sw in used_sw:
                return False
            #Tạo ánh xạ mới bằng cách lưu ký tự từ thứ nhất -> ký tự từ thứ hai trong từ điển
            map_fw_to_sw[char_fw] = char_sw
            #Đánh dấu ký tự từ thứ hai là đã dùng
            used_sw.add(char_sw)
    return True
a, b = map(str, input().split())
print(dang_cau(a, b))
```
